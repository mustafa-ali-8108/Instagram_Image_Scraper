```python
import csv
import os

file =  os.path.join(os.getcwd(),'coffee_culture.csv')

if not os.path.exists(file):
    csv_file = open(file,'w')
    csv_writer = csv.writer(csv_file,lineterminator='\n')
    csv_writer.writerow(['Image Link'])
else:
    csv_file = open(file,'a')
    csv_writer = csv.writer(csv_file,lineterminator='\n')
```
